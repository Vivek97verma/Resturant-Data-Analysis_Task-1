# Analyze the distribution of aggregate ratings and determine the most common rating range.

```python

ratings = df['Aggregate rating']

# Plotting the histogram
plt.figure(figsize=(10, 6))
plt.hist(ratings, bins=10, edgecolor='black')
plt.xlabel('Aggregate Ratings')
plt.ylabel('Frequency')
plt.title('Distribution of Aggregate Ratings')
plt.show()

# Determine the most common rating range
hist, bin_edges = np.histogram(ratings, bins=10)
most_common_range_index = np.argmax(hist)
most_common_range = (bin_edges[most_common_range_index], bin_edges[most_common_range_index + 1])

print(f"The most common rating range is: {most_common_range}")
```

# Calculate the average number of votes received by restaurants.

```python
Avergae_votes=df['Votes'].mean()
print(f" Average number of votes received by resturants is {Avergae_votes:.2f}")
```

# Identify the most common combinations of cuisines in the dataset.

```python
common_cuisines_combination=df['Cuisines'].value_counts().head(10)
print(f' Most common cuisines combination are:')
print(common_cuisines_combination)
```

```python
common_cuisines_combination_df=common_cuisines_combination.reset_index()
common_cuisines_combination_df.columns=['Cuisines','counts']
plt.figure(figsize=(8,5))
sns.barplot(data=common_cuisines_combination_df,x='Cuisines',y='counts',palette='viridis')
plt.title('Most common cuisines combination',fontsize=15)
plt.xticks(rotation=90)
plt.show()
            
```

# Determine if certain cuisine combinations tend to have higher ratings.

```python
Average_cuisines_ratings=df.groupby('Cuisines')['Aggregate rating'].mean().reset_index()
Average_cuisines_ratings.sort_values(by='Aggregate rating',ascending=False)
```

## Task: Geographic Analysis

# Plot the locations of restaurants on a map using longitude and latitude coordinates.

```python
import geopandas as gpd
from shapely.geometry import Point

gdf = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df['Longitude'], df['Latitude']))

# Set the coordinate reference system (CRS) to WGS84 (latitude and longitude)
gdf.set_crs(epsg=4326, inplace=True)

# Plot the locations
fig, ax = plt.subplots(figsize=(10, 10))
gdf.plot(ax=ax, color='blue', markersize=5)

# Add a base map
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
world.plot(ax=ax, color='lightgray', edgecolor='black')

plt.title('Restaurant Locations')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

# Identify any patterns or clusters of restaurants in specific areas.

```python
# Spatial aggregation by city (or any other administrative unit)
restaurants_per_city = gdf.groupby('City').size().reset_index(name='Restaurant Count')

# Optional: You can sort the results to identify cities with the highest number of restaurants
restaurants_per_city.sort_values(by='Restaurant Count', ascending=False, inplace=True)
```
